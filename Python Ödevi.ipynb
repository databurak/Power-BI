## 7.	Yeni Sütun Ekleme

```python
df2 = df.copy()
df2["Temperature_F"] = (df2["Temperature"] * 9/5)+32
df2
```

```python
df2["FeelsLike"] = df2["Temperature"]-(df2["Humidity"]/100)
df2
```
